```
# =======================================================
# ## Conversión a escala de grises
# =======================================================
# Se convierte la imagen creada a una versión en tonos de gris
# para evaluar cómo se representan los niveles de intensidad.

tono_gris = cv2.cvtColor(lienzo, cv2.COLOR_BGR2GRAY)
plt.imshow(tono_gris, cmap="gray")
plt.title("Escala de Grises")
plt.axis("off")
plt.show()
cv2.imwrite("grises.png", tono_gris)
```

```
# =======================================================
# ## Modificación de dimensiones
# =======================================================
# Se amplía la imagen al doble de su tamaño original
# utilizando una función de escalado.

dimension_ampliada = cv2.resize(lienzo, (1000, 1000))
plt.imshow(cv2.cvtColor(dimension_ampliada, cv2.COLOR_BGR2RGB))
plt.title("Redimensionada")
plt.axis("off")
plt.show()
cv2.imwrite("ampliada.png", dimension_ampliada)
```

```
# ## Rotación en diferentes ángulos
# =======================================================
# Se realizan dos rotaciones de la imagen:
# - 45 grados para observar el efecto de inclinación.

(alto, ancho) = lienzo.shape[:2]
punto_central = (ancho // 2, alto // 2)

matriz_rotacion = cv2.getRotationMatrix2D(punto_central, 45, 1.0)
rotada_45 = cv2.warpAffine(lienzo, matriz_rotacion, (ancho, alto))
plt.imshow(cv2.cvtColor(rotada_45, cv2.COLOR_BGR2RGB))
plt.title("Rotada 45 grados")
plt.axis("off")
plt.show()
cv2.imwrite("rotada45.png", rotada_45)


```

```
# ## Rotación en diferentes ángulos
# =======================================================
# Se realizan dos rotaciones de la imagen:
# - 30 grados para observar el efecto de inclinación.
matriz_rotacion_30 = cv2.getRotationMatrix2D(punto_central, 30, 1.0)
rotada_30 = cv2.warpAffine(lienzo, matriz_rotacion_30, (ancho, alto))
plt.imshow(cv2.cvtColor(rotada_30, cv2.COLOR_BGR2RGB))
plt.title("Rotada 30 grados")
plt.axis("off")
plt.show()
cv2.imwrite("rotada30.png", rotada_30)
```

```
# =======================================================
# ## Transformación de perspectiva
# =======================================================
# Se aplica una transformación que altera la perspectiva
# de la imagen, simulando una distorsión visual.

puntos_1 = np.float32([[50, 50], [450, 50], [50, 450], [450, 450]])
puntos_2 = np.float32([[10, 100], [400, 50], [100, 400], [400, 450]])
matriz_perspectiva = cv2.getPerspectiveTransform(puntos_1, puntos_2)
lienzo_perspectiva = cv2.warpPerspective(lienzo, matriz_perspectiva, (500, 500))
plt.imshow(cv2.cvtColor(lienzo_perspectiva, cv2.COLOR_BGR2RGB))
plt.title("Transformación Perspectiva")
plt.axis("off")
plt.show()
cv2.imwrite("perspectiva.png", lienzo_perspectiva)
```

```
# =======================================================
# ## Aplicación de máscara
# =======================================================
# Se emplea una máscara circular que oculta parte de la imagen
# dejando visible únicamente el área interna del círculo.

mascara = np.zeros((500, 500), dtype=np.uint8)
cv2.circle(mascara, (250, 250), 100, 255, -1)
lienzo_mascarado = cv2.bitwise_and(lienzo, lienzo, mask=mascara)
plt.imshow(cv2.cvtColor(lienzo_mascarado, cv2.COLOR_BGR2RGB))
plt.title("Aplicación de Máscara")
plt.axis("off")
plt.show()
cv2.imwrite("mascarado.png", lienzo_mascarado)
```
